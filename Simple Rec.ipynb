```python
def first_take(df):
    #ask for language pref 
    print("Whats your favorite coding language:")
    lang = input()
    
    #ask what size project they want
    print("Do you want a large or small project?")
    size = input()
    #assign variables
    if size == 'small':
        min_issues = 0
        max_issues = 500
    else:
        min_issues = 500
        max_issues = 100000
    
    
    #ask responsibility load
    print("On a scale from 1-3 how much responsibility do you want?")
    resp = int(input())
    
    if resp == 1:
        min_authors = 2000
    elif resp == 2: 
        min_authors = 500
    elif resp == 3:
        min_authors = 2
        
    #create new variable to store preferred projects 
    proj = df.copy()
    
    #filter based on inputs 
    proj = proj[(proj['mainlanguage'] == lang) &
               (proj['open_issues'] >= min_issues) & 
               (proj['open_issues'] <= max_issues) & 
               (proj['numauthors'] >= min_authors)]
    
    
```

```python
first_take(projects)
```
